## Q2. Given a Pandas DataFrame df with columns 'A', 'B', and 'C', write a Python function to re-index the DataFrame with a new index that starts from 1 and increments by 2 for each row.

````python
import pandas as pd

def reindex_by_step(df, step=2):
  """
  Reindexes a pandas DataFrame with a new index starting from 1 and incrementing by the specified step.

  Args:
      df (pd.DataFrame): The DataFrame to reindex.
      step (int, optional): The step value for incrementing the new index. Defaults to 2.

  Returns:
      pd.DataFrame: The reindexed DataFrame.
  """
  # Generate the new index based on the length of the DataFrame and the step value
  new_index = pd.Index(range(1, len(df) * step + 1, step))
  # Reindex the DataFrame using the new index
  return df.set_index(new_index)

# Example usage
data = {'A': [1, 2, 3], 'B': [4, 5, 6], 'C': [7, 8, 9]}
df = pd.DataFrame(data)

df_reindexed = reindex_by_step(df.copy())  # Use a copy to avoid modifying the original DataFrame

print(df_reindexed)
````

## Q7. You have a Pandas DataFrame df that contains a column named 'Email' that contains email addresses in the format 'username@domain.com'. Write a Python function that creates a new column 'Username' in df that contains only the username part of each email address.

The username is the part of the email address that appears before the '@' symbol. For example, if the
email address is 'john.doe@example.com', the 'Username' column should contain 'john.doe'. Your

## Q9. Given a Pandas DataFrame df with a column 'Values', write a Python function to calculate the mean,median, and standard deviation of the values in the 'Values' column.

````python
import pandas as pd

def calculate_statistics(df):
  # Ensure 'Values' column exists
  if 'Values' not in df.columns:
    raise ValueError("DataFrame does not contain a column named 'Values'")

  # Calculate statistics
  statistics = {
      'mean': df['Values'].mean(),
      'median': df['Values'].median(),
      'std': df['Values'].std()
  }
  return statistics

# Example usage
data = {'Values': [10, 5, 15, 20, 8]}
df = pd.DataFrame(data)

stats = calculate_statistics(df.copy())  # Use a copy to avoid modifying the original DataFrame

print(stats)
````

## Q11. You have a Pandas DataFrame df with a column 'Date'. Write a Python function that creates a new column 'Weekday' in the DataFrame. The 'Weekday' column should contain the weekday name (e.g. Monday, Tuesday) corresponding to each date in the 'Date' column.

````python
import pandas as pd

def add_weekday_column(df):
  """
  Adds a new column named 'Weekday' containing weekday names to a DataFrame.

  Args:
      df (pd.DataFrame): The DataFrame containing the 'Date' column.

  Returns:
      pd.DataFrame: The DataFrame with the new 'Weekday' column.
  """
  # Ensure 'Date' column exists
  if 'Date' not in df.columns:
    raise ValueError("DataFrame does not contain a column named 'Date'")

  # Add 'Weekday' column using strftime and day_name
  df['Weekday'] = df['Date'].dt.strftime('%A')
  return df

# Example usage
data = {'Date': pd.to_datetime(['2024-03-05', '2024-03-06', '2024-03-07'])}
df = pd.DataFrame(data)

df = add_weekday_column(df.copy())  # Use a copy to avoid modifying the original DataFrame

print(df)
````

## Q12. Given a Pandas DataFrame df with a column 'Date' that contains timestamps, write a Python function to select all rows where the date is between '2023-01-01' and '2023-01-31'.

````python
import pandas as pd

def filter_by_date_range(df, start_date='2023-01-01', end_date='2023-01-31'):
  """
  Filters a DataFrame based on a date range in the 'Date' column.

  Args:
      df (pd.DataFrame): The DataFrame containing the 'Date' column.
      start_date (str, optional): The starting date of the range (inclusive). Defaults to '2023-01-01'.
      end_date (str, optional): The ending date of the range (inclusive). Defaults to '2023-01-31'.

  Returns:
      pd.DataFrame: The filtered DataFrame containing rows within the date range.
  """
  # Ensure 'Date' column exists
  if 'Date' not in df.columns:
    raise ValueError("DataFrame does not contain a column named 'Date'")

  # Convert dates to datetime format (if necessary)
  df['Date'] = pd.to_datetime(df['Date'])

  # Filter rows based on date range (inclusive)
  filtered_df = df[(df['Date'] >= start_date) & (df['Date'] <= end_date)]
  return filtered_df

# Example usage
data = {'Date': ['2022-12-31', '2023-01-02', '2023-01-15', '2023-02-01']}
df = pd.DataFrame(data)

filtered_df = filter_by_date_range(df.copy())  # Use a copy to avoid modifying the original DataFrame

print(filtered_df)
````
