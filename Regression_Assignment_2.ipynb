## Q1.Explain the concept of R-squared in linear regression models. How is it calculated, and what does it represent?

R Squared is a performance metrics which is used to evaluate the overall performance (total correct prediction) of a regression model.

The value of R-square lies between 0 to 1. Where we get R-square equals 1 when the model perfectly fits the data and there is no difference between the predicted value and actual value.  However, we get R-square equals 0 when the model does not predict any variability in the model and it does not learn any relationship between the dependent and independent variables.

R-Squared (R² or the coefficient of determination) is a statistical measure in a regression model that **determines the proportion of variance in the dependent variable that can be explained by the independent variable**.

In other words, r-squared shows how well the data fit the regression model (the goodness of fit).

**How is R-Squared Calculated:**

r_squared = $1 - (Sum of square Residual / sum of squared total)$

or

$r\_squared = 1- \frac{\sum_{i=1}^n (y_i - h_θ(x))^2}{\sum_{i=1}^n (y_i - y_{mean})^2} $

where,

* The total sum of squares is calculated by summation of squares of perpendicular distance between data points and the average line.

* The residual sum of squares is calculated by the summation of squares of perpendicular distance between data points and the best-fitted line.

## Q2. Define adjusted R-squared and explain how it differs from the regular R-squared.

This leads to the model having high variance if the model has a lot of independent variables.

Formula For Adjusted R-Squared

$Adjusted R^2 = 1 - \frac{(1-R^2) \cdot (n-1)}{n-k-1}$

Here R-square is the value that we calculate using the method explained above

* n is the total number of observations in the data

## Q4. What are RMSE, MSE, and MAE in the context of regression analysis? How are these metrics calculated, and what do they represent?

There are **Three Error Metrics** that are commonly used for evaluating and reporting the performance of a regression model.

1. Mean Squared Error:

$MSE = \frac{\sum_{i=1}^n (y_i - y_{pred})^2 }{n}$

The difference between these two values is sqaured, which has the effect of removing the sign, resulting in a positive error value.

2. Mean Absolute Error:

$MSE = \frac{\sum_{i=1}^n |y_i - y_{pred}| }{n}$

The mean absolute error is calculated as the average of the absolute error values.

Absolute is a mathematical function that simply makes a number positive. Therefore, the dufference betweeen an expected and predicted value may be positive or negative it is forced to be positive when calculating MAE.

3. Root Mean Squared Error:

$RMSE = \sqrt{MSE}$

The Root Mean Squared Error is an extension of the Mean Sqaured Error.

after the square root the units of RMSE are the same as the original units of the target value that is been predicted and that's how the RMSE solves the problem arise in MSE (Mean Squared Error)

## Q5. Discuss the advantages and disadvantages of using RMSE, MSE, and MAE as evaluation metrics in regression analysis.


Root Mean Square Error (RMSE), Mean Squared Error (MSE), and Mean Absolute Error (MAE) are common evaluation metrics in regression analysis. Each metric has its own advantages and disadvantages:

**1. Mean Squared Error (MSE):**

Advantages:
- IT has one local and one global minima.
- MSE is differentiable, making it suitable for optimization algorithms like gradient descent, which is commonly used in machine learning model training.

Disadvantages:
- It is not directly interpretable because the units of MSE are the square of the units of the target variable. This can make it challenging to communicate the error magnitude to non-technical stakeholders.
- MSE can be sensitive to outliers, as squaring the errors amplifies the impact of outliers on the metric. Outliers can disproportionately affect the overall error, potentially leading to misleading conclusions.

**2. Root Mean Square Error (RMSE):**

Advantages:
- RMSE addresses the unit issue of MSE by taking the square root of the MSE, resulting in units that match the original target variable. This makes RMSE more interpretable and easier to communicate.
- Like MSE, RMSE also penalizes large errors more heavily.

Disadvantages:
- It retains the sensitivity to outliers, which can be problematic when dealing with datasets containing extreme values. Outliers can have a significant impact on RMSE.

**3. Mean Absolute Error (MAE):**

Advantages:
- MAE is more robust to outliers compared to MSE and RMSE. It treats all errors equally, making it less influenced by extreme values.
- It provides a direct and interpretable measure of the average absolute error in the same units as the target variable.

Disadvantages:
- MAE does not heavily penalize large errors, which can be a drawback when you want to prioritize the reduction of larger errors in your model.
- It is not differentiable at zero, which can be an issue when using optimization algorithms that rely on gradients.

In summary, the choice of evaluation metric in regression analysis depends on the specific goals of your analysis, the nature of your data, and the impact of outliers. Here are some general guidelines:

- Use MSE or RMSE when you want to give more emphasis to reducing larger errors and when the error distribution is approximately Gaussian (normal).
- Use MAE when you want a metric that is robust to outliers and when you want an easily interpretable error measure.
- Consider using multiple metrics and visualizations to get a comprehensive understanding of your model's performance, especially in situations where the trade-offs between different error metrics are unclear.

##Q8. Discuss the limitations of regularized linear models and explain why they may not always be the best choice for regression analysis.

**Limitations of Regularization**

The limitations of Regularization are:

1. Regularization can make the model too simple and unsuitable for the data if the regularization parameter is too large. This can cause the model to miss important patterns in the data.

2. Regularization assumes that all input variables are of equal importance, which may not be true in some cases. Using domain knowledge to weigh the importance of input variables can help improve model performance.
3.Regularization assumes that the data is linearly separable, which may not be true in some cases. Non-linear models, such as neural networks, maybe a better fit for that data.
4.Regularization assumes that the data are independent and identically distributed (IIDs), which may not be true for some types of data, such as time series or spatial data.

## Q10. You are comparing the performance of two regularized linear models using different types of regularization. Model A uses Ridge regularization with a regularization parameter of 0.1, while Model B uses Lasso regularization with a regularization parameter of 0.5. Which model would you choose as the better performer, and why? Are there any trade-offs or limitations to your choice of regularization method?

The choice between Ridge regularization (L2 regularization) and Lasso regularization (L1 regularization) depends on your specific goals and the characteristics of your data. Both regularization methods have their advantages and limitations, and the choice should align with what you prioritize in your analysis.

Model A (Ridge Regularization with λ = 0.1):

Ridge regularization adds a penalty term to the linear regression model that encourages the coefficients to be small but doesn't force any of them to become exactly zero.

Advantages:
1. Ridge can help prevent overfitting by reducing the magnitude of the regression coefficients.
2. It is often useful when you suspect that multicollinearity (high correlation between predictor variables) might be present in the data, as it tends to shrink correlated coefficients.

Model B (Lasso Regularization with λ = 0.5):

Lasso regularization also adds a penalty term to the linear regression model, but it has the additional effect of shrinking some coefficients all the way to zero, effectively performing feature selection.

Advantages:
1. Lasso is effective for feature selection. If you want a simpler model with fewer predictor variables, Lasso can automatically eliminate less relevant variables by setting their coefficients to zero.
2. It can be particularly useful when dealing with a high-dimensional dataset with many predictors, as it helps to reduce the complexity of the model.

Which model is better depends on your goals:

- If you want a model that retains all your features and simply reduces the magnitude of the coefficients to prevent overfitting, you might prefer Model A with Ridge regularization.

- If you want a more interpretable and potentially simpler model with some variables eliminated from the model, you might prefer Model B with Lasso regularization.

Trade-offs and Limitations:

1. **Feature Selection:** While Lasso can perform feature selection, Ridge doesn't. If you need all the features in your model, Ridge may be a better choice.

2. **Interpretability:** Lasso can lead to more interpretable models by eliminating irrelevant features, but this may come at the cost of some predictive power.

3. **Strength of Regularization (λ):** The choice of the regularization strength (λ) is crucial in both Ridge and Lasso. The optimal λ value can vary, and finding the right value often requires cross-validation.

4. **Data Characteristics:** The effectiveness of Ridge or Lasso can also depend on the specific characteristics of your data. For example, if you have a large number of highly correlated predictors, Ridge might be more suitable.

In summary, your choice between Ridge and Lasso regularization should align with your modeling goals, the importance of feature selection, and the trade-offs you are willing to make between model simplicity and predictive performance. It's often a good practice to experiment with both methods and select the one that best fits your specific situation.
